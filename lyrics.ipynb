```python
nlp = en_core_web_sm.load()
df2 = df.set_index("genre", drop = False)
df2
```

```python
genres = {x for x in df['genre'] if x not in ['Not Available', 'Other']}
genres
```

```python
%%time
placesdict = {genre:[] for genre in genres}

for genre in genres:
    subdf = df2.loc[genre,"lyrics"][:500]
    joined = ' '.join([lyrics for lyrics in subdf if isinstance(lyrics, str)]).replace('\n', ' ')
    #this step takes a lon 
    places = [ent.text for ent in nlp(joined, disable=['parser', 'tagger', 'textcat']).ents if ent.label_ in ['GPE','LOC']]
    placesdict[genre].extend(places)
 

```

```python
from collections import Counter

for genre,places in placesdict.items():
#     print(genre.upper(), ": ", [x for x in places if x in common_places], '\n')
    print(genre.upper(), ": ", Counter(places), '\n')
```
