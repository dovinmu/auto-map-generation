```python
locs = getLocations(text, True)
geojsons = getGeojson(locs)
geojsons = geojsonFilter(geojsons, 'osm')
geojsons = sizeFilter(geojsons)
latLngs = geojsonToLatLng(geojsons)

includesents = True
# there will have been some locations that get an error from the locator API--filter these
filtered_locs = []
for loc in locs:
    if loc[0] in latLngs:
        filtered_locs.append(loc + [latLngs[loc[0]]])

df = pd.DataFrame(filtered_locs, columns=['text', 'label', 'idx', 'sentence', 'coordinates']) if includesents else pd.DataFrame(filtered_locs, columns=['text', 'label', 'idx', 'coordinates'])

gdf = gpd.GeoDataFrame(df, geometry='coordinates')

mapWithLines(gdf, labels=True)
bbox = { 'xmax': -70, 'xmin': -125, 'ymax': 55, 'ymin': 30}
plt.xlim(bbox['xmin'], bbox['xmax'])
plt.ylim(bbox['ymin'], bbox['ymax'])

plt.title("Lewis and Clark Expedition Timeline (article on wikipedia)")
plt.legend()
plt.show()
```
